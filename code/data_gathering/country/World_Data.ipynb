```python
import scrapy 
from scrapy import Selector
import requests
import numpy as np
import pandas as pd
from tqdm import tqdm
import time
```

```python
#Create the selector 
url = 'https://www.worldometers.info/world-population/population-by-country/'
r = requests.get(url)
html = r.content 
sel = Selector(text = html)

path_title = "h1::text"
sel.css(path_title).extract()
```

```python
#Get the country name
names_css = "tbody a ::text"
name = sel.css(names_css).extract()

#Get the population
pop_css = "tbody tr td:nth-of-type(3) ::text"
population = sel.css(pop_css).extract()

den_css = "tbody tr td:nth-of-type(6) ::text"
density = sel.css(den_css).extract()

#Get the size of the country in km2
size_css = "tbody tr td:nth-of-type(7) ::text"
size = sel.css(size_css).extract()
   
#Get the number of migrants
mig_css = "tbody tr td:nth-of-type(8) ::text"
migrants = sel.css(mig_css).extract()

#Get the fertility rate (?)
fert_css = "tbody tr td:nth-of-type(9) ::text"
fertility = sel.css(fert_css).extract()

#Get the median age
age_css = "tbody tr td:nth-of-type(10) ::text"
age = sel.css(age_css).extract()

#Get the percentage of urban population
urb_css = "tbody tr td:nth-of-type(11) ::text"
urban = sel.css(urb_css).extract()

#Get the world share of population
world_css = "tbody tr td:nth-of-type(12) ::text"
worldshare = sel.css(world_css).extract()
   
```

```python
#Delete Kommas from Attributes and convert in int
i=0
while i < 235:
    population[i]=int(population[i].replace("," ,""))
    size[i]=int(size[i].replace("," ,""))
    migrants[i]=migrants[i].replace("," ,"")
    density[i]=int(density[i].replace("," ,""))
    i = i+1
```

```python
world_data = pd.DataFrame()
world_data["Name"]= name
world_data["Population"]= population
world_data["Worldshare"]= worldshare
world_data["Size_in_km2"]= size
world_data["Density"] =density
world_data["UrbanPopulation"]= urban
world_data["Migrants_net"]= migrants
world_data["FertilityRate"]= fertility
world_data["MedianAge"]= age

world_data
```
